### enumerate() built-in function

```python
x=np.random.randint(1,100,10)
x
```

```python
y=enumerate(x)
```

```python
for index,value in y:
    print(index,value)
```

```python
text = "Hello, World!"

for index, char in enumerate(text):
    print(index, char)
```

- enumerate() can be used with array, string, list, tuple.
