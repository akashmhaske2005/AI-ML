### Clean Offer Price Column

```python
df.head()
```

```python
df['Offer price'].unique()
```

```python
df['Offer price']= df['Offer price'].str.replace('₹','')
df['Offer price']= df['Offer price'].str.replace(',','')
```

```python
df['Offer price'].notnull()
```

```python
df.loc[df['Offer price'].notnull(), 'Offer price']
```

```python
df.loc[df['Offer price'].notnull(), 'Offer price']=df.loc[df['Offer price'].notnull(), 'Offer price'].astype(float)
```

```python
df.head()
```

```python
df.dtypes
```

### Clean Actual Price Column

```python
df['Actual price']= df['Actual price'].str.replace('₹','')
df['Actual price']= df['Actual price'].str.replace(',','')
```

```python
df.loc[df['Actual price'].notnull(), 'Actual price']=df.loc[df['Actual price'].notnull(), 'Actual price'].astype(float)
```

```python
df.head()
```

```python
df.info()
```

### Clean Saving Column

```python
df['Saving'].str.split(' ')
```

```python
df['Saving'].str.split(' ').str.get(1)
```

```python
df['Saving']=df['Saving'].str.split(' ').str.get(1)
```

```python
df['Saving'].unique()
```

```python
df['Saving']= df['Saving'].str.replace('₹','')
df['Saving']= df['Saving'].str.replace(',','')
df['Saving']= df['Saving'].replace(['it','Delivery'],np.nan)
```

```python
df.loc[df['Saving'].notnull(), 'Saving']=df.loc[df['Saving'].notnull(), 'Saving'].astype(float)
```

```python
df.head()
```
