### Heat Map

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True, linewidths=0.5)

plt.title("Correlation Heatmap of DataFrame")
plt.show()
```

- annot=True, shows the correlation values inside each cell.

- Use Box Plot, hist plot for numerical columns.
- Use Count, Bar, pie Plot for for categorical columns.
- Use Line and Scatter Plot for Bivariate Analysis of two numerical columns.
- Use categorical columns with hue.
